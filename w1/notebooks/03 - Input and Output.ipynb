## Variables and conversion

You can store values input in variables. All user input is a string, so you have to convert it if you want to use it as a number.

```python
# WHOOPS

a = input("Enter the first number: ")
b = input("Enter the second number: ")
print("The sum is", a + b)
```

How do we fix this?

```python
a = input("Enter the first number: ")
b = input("Enter the second number: ")
print(type(a))
print(type(b))
print("The sum is", int(a) + int(b))
```

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print(type(a))
print(type(b))
print("The sum is", a + b)
```

# Variable names

Variables (along with other things we'll learn about like functions, modules, and methods) are named using "snake case." Snake case is all lowercase letters with underscores between words, like:

* `num_of_animals`
* `current_player`
* `snake_case`
